```python
import pandas as pd # pandas is a data manipulation library
import numpy as np # provides numerical arrays and function to manipulate the arrays efficiently
#import matplotlib.pyplot as plt # data visualisation library
import matplotlib.pyplot as plt
```

```python
# Movies Dataset
mnames = ['movieid','title','genres']
movies_data = pd.read_table('movies.dat', sep='::',header=None, names=mnames, engine= 'python', encoding='latin-1')
print('movie shape:',movies_data.shape)
movies_data.head()
```

```python
# Ratings Datases
rnames = ['userid','movieid','rating','timestamp']
ratings_data = pd.read_table('ratings.dat', sep='::', header=None, names=rnames, engine= 'python', encoding='latin-1')
print('movie shape:',ratings_data.shape)
ratings_data.head()
```

```python
# Merge user and ratings
movie_ratings_data = movies_data.merge (ratings_data, on= 'movieid', how='inner')
movie_ratings_data.head()
```

```python
# Users dataset
unames = ['userid','gender','age','occupation','zipcode']
users_data = pd.read_table('users.dat', sep='::',header=None, names=unames, engine= 'python', encoding='latin-1')
print('movie shape:',users_data.shape)
users_data.head()
```

```python
data = pd.merge(movie_ratings_data , users_data)
data.head(5)
```

### Soal Nomor 1 , Judul film yang paling banyak di sukai wanita

```python
# Cara Yang Salah
# Karena Size Data Mean tidak di ketahui
data_bygender = data.pivot_table(index=['movieid','title'], columns ='gender',values='rating', aggfunc='mean')
data_bygender.head(5)
```

```python
# Movie group by title size
ratings_by_title = data.groupby('title').size()
ratings_by_title
```

```python
# cari index dari title size yang lebih besar dari 250
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles
```
