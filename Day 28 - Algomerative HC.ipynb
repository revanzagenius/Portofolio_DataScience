```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing Dataset
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:,[3,4]].values # Ambil baris kolom 3 dan 4 saja
```

```python
X.shape
```

```python
X
```

## Using The Dendrogram Method to find the optimal numbers clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('DENDROGRAM')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

## Training the Hierarcical Clustering Model on the dataset

```python
from sklearn .cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
y_hc.shape
```

```python
y_hc
```

## Visualising the clustres

```python
# Scatter plot CLUSTER 1 : X = annual income customer kategori 1 (0), y = spending score customer kategori 1 (0)
plt.scatter(X[y_hc == 0, 0],X[y_hc==0,1], s=100, c='red', label='Cluster1')
# Scatter plot CLUSTER 2 : X = annual income customer kategori 2 (1), y = spending score customer kategori 2 (1)
plt.scatter(X[y_hc == 1, 0],X[y_hc==1,1], s=100, c='blue', label='Cluster1')
# Scatter plot CLUSTER 3 : X = annual income customer kategori 3 (2), y = spending score customer kategori 3 (2)
plt.scatter(X[y_hc== 2, 0],X[y_hc==2,1], s=100, c='green', label='Cluster1')
# Scatter plot CLUSTER 4 : X = annual income customer kategori 4 (3), y = spending score customer kategori 4 (3)
plt.scatter(X[y_hc == 3, 0],X[y_hc==3,1], s=100, c='cyan', label='Cluster1')
# Scatter plot CLUSTER 5 : X = annual income customer kategori 5 (4), y = spending score customer kategori 5 (4)
plt.scatter(X[y_hc== 4, 0],X[y_hc==4,1], s=100, c='magenta', label='Cluster1')
# plot info
plt.title('Clusters of Costumers')
plt.xlabel('Anual Income ($k)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
