# Diabetes Prediction

## Importing the libraries

```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score
```

## Importing the dataset

```python
data = pd.read_csv('diabetes_prediction_dataset.csv')
data
```

```python
data.info()
```

## Cleaning the dataset

```python
# check null
data.isnull().sum()
```

```python
# check duplicate
data.duplicated()
```

```python
# drop duplicate
data.drop_duplicates(inplace=True)
```

```python
data.duplicated()
```

## Encoding gender data

```python
data['gender'].value_counts()
```

```python
# do label encoding
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
data['gender'] = label_encoder.fit_transform(data['gender'])
data
```

## Mapping smoking history

```python
# mapping untuk smoking_history
data['smoking_history'].value_counts()
```

```python
smoking_history_maping = {'never': 0, 'No Info': -1, 'current':2, 'not current': 0, 'former' : 1, 'ever': 2}
data['smoking_history'] = data['smoking_history'].map(smoking_history_maping)
data
```

## Checking age data

```python
data.describe()
```

```python
data[data['age'] < 1]
```

```python
data[data['age'] > 1]
```

```python
data['age'].value_counts()
```

```python
data = data[data['age'] > 1]
```

```python
# convert age to int
data['age'] = data['age'].astype(int)
```

```python
data.describe()
```

```python
data.info()
```

## Split train and test set

```python
X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=0)
```

## Feature scaling

```python
sc = preprocessing.StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

## Training and predict the model using Logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

## Training and predict the model using K-NN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
y_pred_knn = knn.predict(X_test)
```

## Linear Linear SVM

```python
from sklearn.svm import SVC
linear_svm = SVC(kernel = 'linear')
linear_svm.fit(X_train, y_train)
```

```python
y_pred_svm = linear_svm.predict(X_test)
```

## Accuracy Score

```python
# logistic regression
accuracy_score(y_test, y_pred_lr)
```

```python
# K-NN
accuracy_score(y_test, y_pred_knn)
```

```python
# Kernel SVM
accuracy_score(y_test, y_pred_svm)
```
