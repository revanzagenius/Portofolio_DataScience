# Random Forest Regression

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Import Dataset

```python
data = pd.read_csv('Position_Salaries.csv')
data
```

```python
X = data.iloc[:,1:-1].values  # independet #diaksih values atau tidak terserah    
y = data.iloc[:,-1].values    #dependen 
```

## Training the Random Forest regression model on the whole dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor= RandomForestRegressor(random_state=0, n_estimators= 10)
regressor.fit(X,y)
```

## PREDIKSI SINGLE VALUE 

```python
regressor.predict([[6.5]])
```

## Visualisasi the Random Forest regression

```python
# garfik scatter
plt.scatter(X,y,color="blue")
# grafik decision tree regression
plt.plot(X,regressor.predict(X),color="red")
plt.title("true or bluff (Random Forest regression)")
plt.xlabel("level posision")
plt.ylabel("salary")
plt.show()
```

```python
X_grid = np.arange(min(X),max(X),0.1)
X_grid = X_grid.reshape((len(X_grid),1)) 
# garfik scatter
plt.scatter(X,y,color="blue")
# grafik decision tree regression
plt.plot(X_grid,regressor.predict(X_grid),color="red")
plt.title("true or bluff (Random Forest regression) ")
plt.xlabel("level posision")
plt.ylabel("salary")
plt.show()
```
