## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
# membantu HRD untuk memastikan seorang kandidat 
# jabatan sebelumya: region manager
# gaji: $165000 per year
# manager HRD: permintaannya benar atau salah?
# jawab dengan model yg bisa memprediksi gaji berdasarkan level jabatannya
# buat prediksi gaji dengan level 6.5 ($165000)
dataset = pd.read_csv('Position_Salaries.csv')
dataset
```

```python
X = dataset.iloc[:,1:-1] # [semua baris, semua kolom kecuali y]
y = dataset.iloc[:,-1] # [semua baris, kolom yg terakhir saja (y)]
```

## Training the linear regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

## Training the polynomial linear regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

## Visualising the linear regression result

```python
# grafik scatter
plt.scatter(X, y, color = "red")
# grafik linier regresi
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or bluff (Linear Regression)')
plt.xlabel('Level position')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial linear regression result

```python
# grafik scatter
plt.scatter(X, y, color = "red")
# grafik linier regresi
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth or bluff (Polynomial Regression)')
plt.xlabel('Level position')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial linear regression result (higher resolution)

```python
X_grid = np.arange(1, 10, 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
X_grid.shape
plt.scatter(X, y, color = "red")
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or bluff (Polynomial Regression) - Higher resolution')
plt.xlabel('Level position')
plt.ylabel('Salary')
plt.show()
```

## Predict single result using simple linear regression

```python
# gaji level 6.5 menurut regresi linear biasa
print(lin_reg.predict([[6.5]]))
```

## Predict single result using polynomial regression

```python
# gaji level 6.5 menurut regresi linear polynimial
print(lin_reg_2.predict(poly_reg.fit_transform([[6.5]])))
```
